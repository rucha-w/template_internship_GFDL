### Sample Reading Data Directly from Dora

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import doralite
```

Get information about an experiment 

```python
doralite.dora_metadata(1188)
```

Load temperature as Pandas Dataframe

```python
df = doralite.global_mean_data(1188, 'globalAveAtmos')
```

```python
df.columns
```

```python
df.t_ref.plot()
```

```python
df_smoothed = df.smooth(20)
```

```python
df_smoothed.t_ref.plot()
```

```python
df_detrend = df.detrend()
```

```python
df_detrend.t_ref.plot()
```

```python
df_detrend_3 = df.detrend(order=3)
```

```python
df_detrend_3.t_ref.plot()
```

```python
fig = plt.figure()
ax = plt.subplot(1,1,1)
df.t_ref.plot(ax=ax, color = '#a3a3a3', linewidth=0.5, label='raw')
df_smoothed.t_ref.plot(ax=ax, color='red', linewidth=2.0, label='smoothed')
ax.set_ylabel('degrees K')
ax.set_title(' Global Historical Temperature Curve', fontsize=22)
plt.legend()
```
